```python
import os
import pandas as pd
```

```python
def processar_arquivos(pasta):
    
    colunas = ["Ano", "Concurso"] + list(range(1, 51)) + [f"{i}e" for i in range(1, 13)]
    df_final = pd.DataFrame(columns=colunas)
    
    for arquivo in os.listdir(pasta):
        if arquivo.endswith("_eurom.csv"):
            ano = arquivo.split("_")[0]
            caminho = os.path.join(pasta, arquivo)
            df = pd.read_csv(caminho)
            
            df["Ano"] = int(ano)
            
            col_numeros = df.columns[2:7]
            col_estrelas = df.columns[7:9] 
            
            for i in range(1, 51):
                df[i] = df.apply(lambda row: 1 if str(i) in row[col_numeros].astype(str).values else 0, axis=1)
            
            for i in range(1, 13):
                df[f"{i}e"] = df.apply(lambda row: 1 if str(i) in row[col_estrelas].astype(str).values else 0, axis=1)
            
            df_final = pd.concat([df_final, df[colunas]], ignore_index=True)
            del df 
    
    return df_final
```

```python

# Processar os arquivos e salvar o resultado
pasta_dados = "dados"
df_sorteios = processar_arquivos(pasta_dados)
df_sorteios.to_csv("dados/sorteios_eurom.csv", index=False)
print("Arquivo sorteios_eurom.csv salvo com sucesso!")
```
